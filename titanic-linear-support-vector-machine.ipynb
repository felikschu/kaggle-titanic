```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
def dataclean_train(dataname):
    train_data = pd.read_csv(dataname)
    train_data1=train_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]
    train_data1['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
    
    from sklearn.preprocessing import MinMaxScaler

    scaler = MinMaxScaler()
    train_data1[['Age', 'Fare']] = scaler.fit_transform(train_data[['Age', 'Fare']])
    train_data1 = train_data1.fillna(train_data1.median())
    return train_data1

def dataclean_test(dataname):
    train_data = pd.read_csv(dataname)
    train_data1=train_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
    train_data1['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
    
    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    train_data1[['Age', 'Fare']] = scaler.fit_transform(train_data[['Age', 'Fare']])
    train_data1 = train_data1.fillna(train_data1.median())
    return train_data1



    

    
```

```python
train_data1=dataclean_train("/kaggle/input/titanic/train.csv")
test_data1=dataclean_test("/kaggle/input/titanic/test.csv")
```

```python
test_data1
```

```python
X = train_data1.drop('Survived', axis=1)
y = train_data1['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
# 线性支持向量机
model_SVC = SVC(kernel='linear')
model_SVC.fit(X_train, y_train)
```

```python
# 预测
y_pred = model_SVC.predict(X_test)

# 计算准确率
accuracy1 = accuracy_score(y_test, y_pred)

print(f"Accuracy_SVC（验证集）: {accuracy1:.2f}")
```

```python
from sklearn.linear_model import LogisticRegression
model_Log = LogisticRegression()
model_Log.fit(X_train, y_train)

# 预测
y_pred = model_Log.predict(X_test)

# 计算准确率
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy_Log(验证集): {accuracy:.2f}")
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42)

# 训练决策树模型
model.fit(X_train, y_train)

# 预测
y_pred = model.predict(X_test)

# 计算准确率
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# 随机森林
from sklearn.ensemble import RandomForestClassifier

# 创建随机森林分类器
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 预测和评估
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.naive_bayes import GaussianNB

# 创建朴素贝叶斯分类器
model = GaussianNB()
model.fit(X_train, y_train)

# 预测和评估
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# 梯度提升树
from sklearn.ensemble import GradientBoostingClassifier

# 创建梯度提升分类器
model = GradientBoostingClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 预测和评估
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
